## Train multiple classification models using sklearn

```python
import pandas as pd
import numpy as np
df = pd.read_csv("banknote_authentication.csv", header=None)
df.columns = ["variance", "skewness", "curtosis", "entropy", "class"]
df.head()
```

```python
X = df.drop("class", axis=1)
y = df["class"]

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42
)

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit trên train
scaler.fit(X_train)

# transform
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled
```

```python

print("Mean:", X_train_scaled.mean(axis=0))
print("Std:", X_train_scaled.std(axis=0))
```

```python
df.isnull().sum()
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X.columns)
```

```python
df.dtypes
```

```python
import joblib

joblib.dump(X_train_scaled, "X_train.pkl")
joblib.dump(X_test_scaled, "X_test.pkl")
joblib.dump(y_train, "y_train.pkl")
joblib.dump(y_test, "y_test.pkl")
```

```python
import numpy as np
import pandas as pd

from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
print("Models to be used:")
print("1. AdaBoost")
print("2. Gradient Boosting")
print("3. LDA")
print("4. MLP")
print("5. Logistic Regression")
```

```python
model_ada = AdaBoostClassifier()
model_gb = GradientBoostingClassifier()
model_lda = LinearDiscriminantAnalysis()
model_mlp = MLPClassifier(max_iter=300)  # tránh lỗi chưa hội tụ
model_lr = LogisticRegression(max_iter=1000)

print("Models initialized successfully!")
```

```python
print("Training models...")

model_ada.fit(X_train_scaled, y_train)
model_gb.fit(X_train_scaled, y_train)
model_lda.fit(X_train_scaled, y_train)
model_mlp.fit(X_train_scaled, y_train)
model_lr.fit(X_train_scaled, y_train)   

print("Training completed!")
```

```python
y_pred_ada = model_ada.predict(X_test_scaled)
y_pred_gb = model_gb.predict(X_test_scaled)
y_pred_lda = model_lda.predict(X_test_scaled)
y_pred_mlp = model_mlp.predict(X_test_scaled)
y_pred_lr = model_lr.predict(X_test_scaled)

print("Prediction completed!")

print("\nSample predictions (first 10):")
print("AdaBoost:", y_pred_ada[:10])
print("GradientBoosting:", y_pred_gb[:10])
print("LDA:", y_pred_lda[:10])
print("MLP:", y_pred_mlp[:10])
print("LogisticRegression:", y_pred_lr[:10])
```

```python
results = {
    "AdaBoost": y_pred_ada,
    "GradientBoosting": y_pred_gb,
    "LDA": y_pred_lda,
    "MLP": y_pred_mlp,
    "LogisticRegression": y_pred_lr
}

print("Results saved successfully!")
```

```python
models = {
    "AdaBoost": model_ada,
    "GradientBoosting": model_gb,
    "LDA": model_lda,
    "MLP": model_mlp,
    "LogisticRegression": model_lr
}

print("Models stored successfully!")
```

```python
print("\nModel Accuracy:")

for name, model in models.items():
    y_pred = model.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name}: {acc:.4f}")
```
