## Bài 5. Tính tổng các số chẵn
### Nhập một số nguyên n. Hãy tính tổng các số chẵn từ 1 đến n.

```python
# Nhập dữ liệu
n = int(input("Moi ban nhap so nguyen n: "))
# Xuất dữ liệu
s = 0
i = 1

while i <= n:
    if i % 2 == 0:
        s += i
    i += 1

print(f'Tong cac so chan tu 1 den {n} la {s}.')
```
