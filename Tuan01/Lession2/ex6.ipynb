## Bài 6. Kiểm tra số nguyên tố
### Nhập vào một số nguyên dương n. Hãy kiểm tra xem n có phải là nguyên tố không?

Lưu ý: n là số nguyên tố khi và chỉ khi không tồn tại số nguyên từ $2$ đến $\sqrt(n)$ mà chia
hết cho n.

```python
# Nhập dữ liệu
n = int(input("Moi ban nhap so nguyen n: "))
# Xử lý

d = 0

if n < 2:
    lant = False
else:
    for i in range(2, int(n**0.5) + 1):
        if n%i == 0:
            lant = False
            d += 1
            break
    if d == 0:
        lant = True

# Xuất dữ liệu
if lant == True:
    print(f'{n} la so nguyen to.')
else:
    print(f'{n} khong la so nguyen to.')
```
